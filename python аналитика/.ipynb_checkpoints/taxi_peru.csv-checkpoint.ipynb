```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
taxi = pd.read_csv('taxi_peru.csv', sep=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
```

```python
taxi
```

```python
source_df = taxi.groupby('source', as_index=False).agg({'user_id': 'count'}).sort_values('user_id', ascending=False)
```

```python
#source_df['percent'] = round(source_df.user_id / source_df.user_id.sum() * 100)
```

```python
source_df
```

```python
ax = sns.barplot(x='source', y='user_id', data=source_df)
```

```python
icon_df = taxi.groupby('icon', as_index=False).agg({'user_id': 'count'}).sort_values('user_id', ascending=False)
```

```python
icon_df
```

```python
ax1 = sns.countplot(taxi['icon'])
```
