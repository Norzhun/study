```python
import pandas as pd
```

```python
bookings = pd.read_csv('bookings.csv', sep=';')
```

```python
bookings.head()
```

```python
bookings.shape
```

```python
bookings.dtypes
```

```python
bookings = bookings.rename(columns = dict(zip(bookings.columns, [x.lower().replace(' ', '_') for x in bookings.columns])))
```

```python
res2 = bookings.query('is_canceled == 0').groupby(['country']).aggregate({'is_canceled': 'count'}).sort_values('is_canceled', ascending=False)
```

```python
res2
```

```python
bookings.query('hotel == "City Hotel" or hotel == "Resort Hotel"').groupby(['hotel']).aggregate({'stays_total_nights': 'mean'})
```

```python
round(4.318547, 2)
```

```python
bookings.query('assigned_room_type != reserved_room_type').aggregate({'is_canceled': 'count'})
```

```python
bookings.query('arrival_date_year == 2016').groupby('arrival_date_month').aggregate({'arrival_date_year': 'count'}).sort_values('arrival_date_year', ascending=False)
```

```python
bookings.query('arrival_date_year == 2017').groupby('arrival_date_month').aggregate({'arrival_date_year': 'count'}).sort_values('arrival_date_year', ascending=False)
```

```python
bookings.query('arrival_date_year in [2015, 2016, 2017] and is_canceled == 1 and hotel == "City Hotel"').groupby(['arrival_date_year', 'arrival_date_month']).aggregate({'is_canceled': 'sum'}).sort_values('is_canceled', ascending=False)
```

```python
bookings['total_kids'] = bookings.children + bookings.babies
```
